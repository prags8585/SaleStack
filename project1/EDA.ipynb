```python
purchase_summary = pd.read_sql_query("""SELECT 
    p.VendorNumber AS SaleNumber,
    p.VendorName AS SaleName,
    p.Brand,
    p.Description,
    p.PurchasePrice AS CostPrice,
    pp.Price as ActualPrice,
    pp.Volume AS Quantity,
    SUM(p.Quantity) AS TotalQuantityPurchased,
    SUM(p.Dollars) AS TotalDollarsPurchased
FROM purchases p
JOIN purchase_cost pp
    ON p.Brand = pp.Brand
WHERE p.PurchasePrice > 0
GROUP BY p.VendorNumber,p.VendorName, p.Brand, p.Description, p.PurchasePrice, pp.Price, pp.Volume
ORDER BY TotalDollarsPurchased
""",connection)
purchase_summary
```

```python
sales_summary = pd.read_sql_query("""SELECT 
    VendorNo AS SaleNumber,
    Brand,
    SUM(SalesQuantity) AS TotalQuantitySales,
    SUM(SalesDollars) AS TotalDollarsSales,
    SUM(SalesPrice) AS TotalPriceSales,
    SUM(ExciseTax) AS TotalTax
FROM sales
GROUP BY VendorNo, Brand""",connection)
sales_summary
```

```python
# merging all the summary table to create an aggregated summary table for further analysis
sales_stack_summary = pd.read_sql_query("""WITH FreightSummary AS (
    SELECT 
        VendorNumber AS SaleNumber, 
        SUM(Freight) AS FreightCost 
    FROM sales_invoices 
    GROUP BY VendorNumber
), 

PurchaseSummary AS (
    SELECT 
        p.VendorNumber AS SaleNumber,
        p.VendorName AS SaleName,
        p.Brand,
        p.Description,
        p.PurchasePrice AS CostPrice,
        pp.Price as ActualPrice,
        pp.Volume AS Quantity,
        SUM(p.Quantity) AS TotalQuantityPurchased,
        SUM(p.Dollars) AS TotalDollarsPurchased
    FROM purchases p
    JOIN purchase_cost pp
        ON p.Brand = pp.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY p.VendorNumber, p.VendorName, p.Brand, p.Description, p.PurchasePrice, pp.Price, pp.Volume
), 

SalesSummary AS (
    SELECT 
        VendorNo AS SaleNumber,
        Brand,
        SUM(SalesQuantity) AS TotalQuantitySales,
        SUM(SalesDollars) AS TotalDollarsSales,
        SUM(SalesPrice) AS TotalPriceSales,
        SUM(ExciseTax) AS TotalTax
    FROM sales
    GROUP BY VendorNo, Brand
) 

SELECT 
    ps.SaleNumber,
    ps.SaleName,
    ps.Brand,
    ps.Description,
    ps.CostPrice,
    ps.ActualPrice,
    ps.Quantity,
    ps.TotalQuantityPurchased,
    ps.TotalDollarsPurchased,
    ss.TotalQuantitySales,
    ss.TotalDollarsSales,
    ss.TotalPriceSales,
    ss.TotalTax AS TotalTax,
    fs.FreightCost AS CostFreight
FROM PurchaseSummary ps
LEFT JOIN SalesSummary ss 
    ON ps.SaleNumber = ss.SaleNumber 
    AND ps.Brand = ss.Brand
LEFT JOIN FreightSummary fs 
    ON ps.SaleNumber = fs.SaleNumber
ORDER BY ps.TotalDollarsPurchased DESC""", connection)
```

## data cleanning

```python
sales_stack_summary.dtypes
```

```python
sales_stack_summary.isnull().sum()
```

```python
sales_stack_summary['Quantity'].unique()
```

```python
sales_stack_summary['SaleName'].unique()
```

```python
# creating new columns for better analysis
sales_stack_summary['GrossProfit'] = sales_stack_summary['TotalDollarsSales'] - sales_stack_summary['TotalDollarsPurchased']
sales_stack_summary['ProfitMargin'] = (sales_stack_summary['GrossProfit'] / sales_stack_summary['TotalDollarsSales'])*100
sales_stack_summary['Stocks'] = sales_stack_summary['TotalQuantitySales'] / sales_stack_summary['TotalQuantityPurchased']
sales_stack_summary['PurchaseToSalesRatio'] = sales_stack_summary['TotalDollarsSales'] / sales_stack_summary['TotalDollarsPurchased']

```

```python
sales_stack_summary.columns
```

```python
# changing datatype to float
sales_stack_summary['Quantity'] = sales_stack_summary['Quantity'].astype('float')

# filling missing value with 0
sales_stack_summary.fillna(0,inplace = True)

# removing spaces from categorical columns
sales_stack_summary['SaleName'] = sales_stack_summary['SaleName'].str.strip()
sales_stack_summary['Description'] = sales_stack_summary['Description'].str.strip()
```
